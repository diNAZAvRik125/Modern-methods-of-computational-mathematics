```python
import numpy as np
import matplotlib.pyplot as plt
```

##    ##

```python
a_0 = 1
a_1 = 2e-3
w_0 = 5.1
w_1 = 25.5
T = 2*np.pi
```

##   ##

```python
n = 100
t = np.linspace(0, T, n)
f = a_0 * np.sin(w_0 * t) + a_1 * np.sin(w_1 * t) 
plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(t, f)
plt.show()
```

##    ##

```python
fft = np.fft.fft(f)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(t, fft)
plt.show()
```

```python
power_spectrum_rect = np.abs(fft)**2 / len(f)
plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(t, power_spectrum_rect)
plt.show()
```

##      ##

```python
window = np.hanning(n)
f_hann = f * window
fft_hann = np.fft.fft(f_hann)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(t, fft_hann)
plt.show()
```

```python
power_spectrum_hann = np.abs(fft_hann)**2 / len(f)
plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(t, power_spectrum_hann)
plt.show()
```

```python
freq = np.fft.fftfreq(n, d=(T/n))

plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2, 1, 1)
ax2 = plt.subplot(2, 1, 2)

ax1.set_title(' ')
ax1.set_yscale('log')
ax1.set_xlabel('')
ax1.set_ylabel('. ')
ax1.plot(freq, np.abs(fft) ** 2)
ax2.set_title(' ')
ax2.set_yscale('log')
ax2.set_xlabel('')
ax2.set_ylabel('. ')
ax2.plot(freq, np.abs(fft_hann) ** 2)

ax2.grid()
plt.tight_layout()
plt.show()
```
